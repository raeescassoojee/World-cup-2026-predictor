```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

sns.set_style('whitegrid')
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('results.csv')
print(f"Dataset: {df.shape[0]} matches, {df.shape[1]} columns")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
print(f"Unique teams: {df['home_team'].nunique()}")
print(f"\nColumns: {df.columns.tolist()}")
df.head(10)
```

```python
print("Missing values:")
print(df.isnull().sum())
print(f"\nTournament types:\n{df['tournament'].value_counts().head(15)}")
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['goal_diff'] = df['home_score'] - df['away_score']
df['result'] = df['goal_diff'].apply(lambda x: 'Home Win' if x > 0 else ('Away Win' if x < 0 else 'Draw'))
df['total_goals'] = df['home_score'] + df['away_score']
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Goals per year trend
df.groupby('year')['total_goals'].mean().plot(ax=axes[0,0], title='Avg Goals Per Match Over Time')

# Home advantage
df['result'].value_counts().plot.bar(ax=axes[0,1], title='Home Win vs Away Win vs Draw')

# Most matches played (top 15)
all_teams = pd.concat([df['home_team'], df['away_team']])
all_teams.value_counts().head(15).plot.barh(ax=axes[1,0], title='Most Matches Played')

# Goals distribution
df['total_goals'].plot.hist(bins=15, ax=axes[1,1], title='Goals Per Match Distribution')

plt.tight_layout()
plt.show()
```

```python
# Not all matches are equal â a World Cup final matters more than a friendly
# We assign weights so our model values competitive matches higher

tournament_weight = {
    'FIFA World Cup': 1.0,
    'Copa AmÃ©rica': 0.85,
    'UEFA Euro': 0.85,
    'African Cup of Nations': 0.85,
    'AFC Asian Cup': 0.85,
    'CONCACAF Gold Cup': 0.80,
    'FIFA World Cup qualification': 0.75,
    'UEFA Euro qualification': 0.70,
    'African Cup of Nations qualification': 0.70,
    'AFC Asian Cup qualification': 0.70,
    'UEFA Nations League': 0.70,
    'CONCACAF Nations League': 0.65,
    'Friendly': 0.4
}

# Any tournament not in our dictionary gets a default weight of 0.5
df['match_weight'] = df['tournament'].map(tournament_weight).fillna(0.5)

print("Weight distribution:")
print(df['match_weight'].value_counts().sort_index())
```

```python
def get_team_stats(df, team, date, n_matches=30):
    """
    For a given team and date, look back at their last n matches
    and calculate form stats. This is what the ML model will use.
    """
    # Get all matches where this team played (home or away), before this date
    team_matches = df[
        ((df['home_team'] == team) | (df['away_team'] == team)) &
        (df['date'] < date)
    ].tail(n_matches)  # take only the last n matches
    
    if len(team_matches) < 5:
        # Not enough history â return None so we skip this match later
        return None
    
    stats = {}
    stats['matches_played'] = len(team_matches)
    
    wins = 0
    draws = 0
    losses = 0
    goals_scored = 0
    goals_conceded = 0
    weighted_wins = 0
    weighted_points = 0
    
    for _, match in team_matches.iterrows():
        weight = match['match_weight']
        
        if match['home_team'] == team:
            gs = match['home_score']
            gc = match['away_score']
        else:
            gs = match['away_score']
            gc = match['home_score']
        
        goals_scored += gs
        goals_conceded += gc
        
        if gs > gc:
            wins += 1
            weighted_wins += weight
            weighted_points += 3 * weight
        elif gs == gc:
            draws += 1
            weighted_points += 1 * weight
        else:
            losses += 1
    
    stats['win_rate'] = wins / len(team_matches)
    stats['draw_rate'] = draws / len(team_matches)
    stats['loss_rate'] = losses / len(team_matches)
    stats['avg_goals_scored'] = goals_scored / len(team_matches)
    stats['avg_goals_conceded'] = goals_conceded / len(team_matches)
    stats['goal_diff_avg'] = stats['avg_goals_scored'] - stats['avg_goals_conceded']
    stats['weighted_win_rate'] = weighted_wins / len(team_matches)
    stats['weighted_points_avg'] = weighted_points / len(team_matches)
    
    return stats

# Quick test â let's see Brazil's stats going into 2024
test = get_team_stats(df, 'Brazil', pd.Timestamp('2024-01-01'), n_matches=30)
print("Brazil's last 30 matches before 2024:")
for k, v in test.items():
    print(f"  {k}: {v:.3f}")
```

```python
def get_head_to_head(df, team1, team2, date, n_matches=10):
    """
    Get the head-to-head record between two specific teams.
    This captures rivalries and matchup-specific patterns.
    """
    h2h = df[
        (
            ((df['home_team'] == team1) & (df['away_team'] == team2)) |
            ((df['home_team'] == team2) & (df['away_team'] == team1))
        ) &
        (df['date'] < date)
    ].tail(n_matches)
    
    if len(h2h) == 0:
        return {'h2h_win_rate': 0.5, 'h2h_goal_diff': 0, 'h2h_matches': 0}
    
    team1_wins = 0
    team1_goal_diff = 0
    
    for _, match in h2h.iterrows():
        if match['home_team'] == team1:
            diff = match['home_score'] - match['away_score']
        else:
            diff = match['away_score'] - match['home_score']
        
        team1_goal_diff += diff
        if diff > 0:
            team1_wins += 1
    
    return {
        'h2h_win_rate': team1_wins / len(h2h),
        'h2h_goal_diff': team1_goal_diff / len(h2h),
        'h2h_matches': len(h2h)
    }

# Test â Brazil vs Argentina head to head before 2024
h2h = get_head_to_head(df, 'Brazil', 'Argentina', pd.Timestamp('2024-01-01'))
print("Brazil vs Argentina H2H before 2024:")
for k, v in h2h.items():
    print(f"  {k}: {v:.3f}" if isinstance(v, float) else f"  {k}: {v}")
```

```python
# We'll only use matches from 2000 onwards for training
# Older matches are less relevant to modern football
# We also need each team to have enough history (handled by get_team_stats returning None)

from tqdm.notebook import tqdm  # progress bar so we can see how long it takes

df_modern = df[df['year'] >= 2000].copy()
print(f"Matches from 2000 onwards: {len(df_modern)}")

features_list = []

for idx, match in tqdm(df_modern.iterrows(), total=len(df_modern), desc="Building features"):
    
    date = match['date']
    home = match['home_team']
    away = match['away_team']
    
    # Get each team's recent form
    home_stats = get_team_stats(df, home, date, n_matches=30)
    away_stats = get_team_stats(df, away, date, n_matches=30)
    
    # Skip if either team doesn't have enough history
    if home_stats is None or away_stats is None:
        continue
    
    # Get head to head
    h2h = get_head_to_head(df, home, away, date, n_matches=10)
    
    # Build the feature row â the difference between home and away team stats
    # plus some absolute values the model might find useful
    row = {
        # Target variable
        'result': match['result'],
        
        # Match info (not used as features, but useful for reference)
        'date': date,
        'home_team': home,
        'away_team': away,
        'tournament': match['tournament'],
        'neutral': match['neutral'],
        
        # Home team stats
        'home_win_rate': home_stats['win_rate'],
        'home_goals_scored': home_stats['avg_goals_scored'],
        'home_goals_conceded': home_stats['avg_goals_conceded'],
        'home_goal_diff': home_stats['goal_diff_avg'],
        'home_weighted_points': home_stats['weighted_points_avg'],
        
        # Away team stats
        'away_win_rate': away_stats['win_rate'],
        'away_goals_scored': away_stats['avg_goals_scored'],
        'away_goals_conceded': away_stats['avg_goals_conceded'],
        'away_goal_diff': away_stats['goal_diff_avg'],
        'away_weighted_points': away_stats['weighted_points_avg'],
        
        # Differences (home advantage relative to away)
        'win_rate_diff': home_stats['win_rate'] - away_stats['win_rate'],
        'goal_diff_diff': home_stats['goal_diff_avg'] - away_stats['goal_diff_avg'],
        'weighted_points_diff': home_stats['weighted_points_avg'] - away_stats['weighted_points_avg'],
        'goals_scored_diff': home_stats['avg_goals_scored'] - away_stats['avg_goals_scored'],
        'goals_conceded_diff': home_stats['avg_goals_conceded'] - away_stats['avg_goals_conceded'],
        
        # Head to head
        'h2h_win_rate': h2h['h2h_win_rate'],
        'h2h_goal_diff': h2h['h2h_goal_diff'],
        'h2h_matches': h2h['h2h_matches'],
        
        # Match context
        'is_neutral': int(match['neutral']),
        'match_weight': match['match_weight'],
    }
    
    features_list.append(row)

df_features = pd.DataFrame(features_list)
print(f"\nTraining dataset built: {df_features.shape[0]} matches with {df_features.shape[1]} columns")
print(f"\nResult distribution:")
print(df_features['result'].value_counts())
print(f"\nSample of features:")
df_features.head()
```
